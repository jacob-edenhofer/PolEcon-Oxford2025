```python
# import relevant libraries 
import os
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# set path to figures directory 
wd = os.getcwd()
img_dir = os.path.join(wd, 'Figures')
if not os.path.exists(img_dir):
    os.makedirs(img_dir)
```

```python
# Define labels for the three sets
labels = {
    '100': 'Broad individual/\ngroup-based transfers',
    '010': 'Broad place-based\ncorruption',
    '001': 'Targeted individual/group\n-based corruption',
    '110': 'Broad place-based\npolicies (e.g. large-scale IP)',
    '101': 'Clientelism &\ntargeted ind./\ngroup-based benefits',
    '011': 'Regional\ncorruption and\npork-barrel\npolitics',
    '111': 'Pork-barrel politics',
}

# Create the Venn diagram
plt.figure(figsize=(10, 8))
venn = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels=('Legal', 'Geographic', 'Selective'))

# Annotate the regions manually
for region_code, label in labels.items():
    subset = venn.get_label_by_id(region_code)
    if subset:
        subset.set_text(label)
        subset.set_fontsize(8)
plt.savefig(os.path.join(img_dir, 'venn_diagram_v1.png'), dpi=300, bbox_inches='tight')
# plt.title("Conceptual Venn Diagram of Distributive Politics")
plt.show()
```

```python
# Define updated labels with pork-barrel notation
labels = {
    '100': 'Universalistic\nInstitutionalised Transfers',
    '010': 'Informal\nRegional Transfers',
    '001': 'Informal Group-Based\nTransfers',
    '110': 'Institutionalised\nRegional Programmes',
    '101': 'Institutionalised Group-\nBased Programmes',
    '011': 'Informal Group-Based\nRegional Patronage\n(Pork-Barrel Politics)',
    '111': 'Institutionalised Group-Based\nRegional Redistribution\n(Pork-Barrel Politics)',
}

# Create the Venn diagram
plt.figure(figsize=(10, 8))
venn = venn3(
    subsets=(1, 1, 1, 1, 1, 1, 1),
    set_labels=('Institutionalised', 'Geographic', 'Group-Based')
)

# Annotate the regions manually
for region_code, label in labels.items():
    subset = venn.get_label_by_id(region_code)
    if subset:
        subset.set_text(label)
        subset.set_fontsize(6)  # Reduced font size for better fit

# Save and show the figure
plt.savefig(os.path.join(img_dir, 'venn_diagram_v2.png'), dpi=300, bbox_inches='tight')
plt.show()
```
