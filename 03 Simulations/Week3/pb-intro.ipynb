```python
# import relevant libraries 
import os
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# set path to figures directory 
wd = os.getcwd()
img_dir = os.path.join(wd, 'Figures')
if not os.path.exists(img_dir):
    os.makedirs(img_dir)
```

```python
# Define labels for the three sets
labels = {
    '100': 'Broad individual/\ngroup-based transfers',
    '010': 'Broad place-based\ncorruption',
    '001': 'Targeted individual/group\n-based corruption',
    '110': 'Broad place-based\npolicies (e.g. large-scale IP)',
    '101': 'Clientelism &\ntargeted ind./\ngroup-based benefits',
    '011': 'Regional\ncorruption',
    '111': 'Pork-barrel politics',
}

# Create the Venn diagram
plt.figure(figsize=(10, 8))
venn = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels=('Legal', 'Geographic', 'Selective'))

# Annotate the regions manually
for region_code, label in labels.items():
    subset = venn.get_label_by_id(region_code)
    if subset:
        subset.set_text(label)
        subset.set_fontsize(8)
plt.savefig(os.path.join(img_dir, 'venn_diagram.png'), dpi=300, bbox_inches='tight')
# plt.title("Conceptual Venn Diagram of Distributive Politics")
plt.show()
```
