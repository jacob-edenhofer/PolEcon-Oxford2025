```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Define alpha values from 0.05 to 0.95
alpha_vals = np.linspace(0.05, 0.95, 19)

# Ideal points
x_L, x_R = 0.2, 1.0
y_L, y_R = 1.0, 0.2

# Compute x*(alpha) and y*(alpha)
x_star = alpha_vals * x_L + (1 - alpha_vals) * x_R
y_star = (1 - alpha_vals) * y_L + alpha_vals * y_R

# Plotting
# Create a two-panel plot to separate x* and y* clearly
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# Panel 1: x*(alpha)
ax1.plot(alpha_vals, x_star, color='darkorange', marker='o', label=r"$x^*(\alpha)$")
ax1.axhline(y=x_L, color='darkorange', linestyle='--', linewidth=0.8, label=r"$x_L = 0.2$")
ax1.axhline(y=x_R, color='darkorange', linestyle=':', linewidth=0.8, label=r"$x_R = 1.0$")
ax1.set_ylabel(r"$x^*(\alpha)$")
ax1.set_title(r"Policy Outcome $x^*(\alpha)$ as Salience $\alpha$ Varies")
ax1.legend()
ax1.grid(True)

# Panel 2: y*(alpha)
ax2.plot(alpha_vals, y_star, color='steelblue', marker='s', label=r"$y^*(\alpha)$")
ax2.axhline(y=y_L, color='steelblue', linestyle='--', linewidth=0.8, label=r"$y_L = 1.0$")
ax2.axhline(y=y_R, color='steelblue', linestyle=':', linewidth=0.8, label=r"$y_R = 0.2$")
ax2.set_xlabel(r"Salience weight $\alpha$ (for $L$ on $x$)")
ax2.set_ylabel(r"$y^*(\alpha)$")
ax2.set_title(r"Policy Outcome $y^*(\alpha)$ as Salience $\alpha$ Varies")
ax2.legend()
ax2.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
