```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set(style="whitegrid")
```

```python
# Parameters
x_ideal = 0.3
x_R = 0.7
v_P = 0.2
v_R = 0
lambdas = [0.0, 0.25, 0.5, 0.75, 1.0]
colors = ["navy", "teal", "green", "orange", "red"]

# Domain of positions
x_vals = np.linspace(0, 1, 500)

# Utility function
def party_utility(x_P, x_ideal, x_R, v_P, v_R, ):
    V_P = (x_P + x_R + (v_P - v_R)) / 2
    return  * V_P - (1 - ) * (x_P - x_ideal)**2

# Plotting
plt.figure(figsize=(9, 6))

# Plot utility curves
for , color in zip(lambdas, colors):
    U_vals = [party_utility(x, x_ideal, x_R, v_P, v_R, ) for x in x_vals]
    plt.plot(x_vals, U_vals, label=fr"$\lambda = {}$", color=color)

# Add key position markers
plt.scatter([x_ideal], [0], color='blue', label="Party P's Ideal", zorder=5)
plt.scatter([x_R], [0], color='darkorange', label="Party R's Ideal", zorder=5)
plt.scatter([0.5], [0], color='black', label="Median Voter", zorder=5)

# Annotations instead of vertical lines
plt.annotate("P's Ideal Point", xy=(x_ideal, 0), xytext=(x_ideal-0.05, -0.15),
             arrowprops=dict(arrowstyle='->', color='blue'), color='blue')
plt.annotate("R's Ideal Point", xy=(x_R, 0), xytext=(x_R-0.1, -0.25),
             arrowprops=dict(arrowstyle='->', color='darkorange'), color='darkorange')
plt.annotate("Median Voter", xy=(0.5, 0), xytext=(0.5-0.08, 0.05),
             arrowprops=dict(arrowstyle='->', color='black'), color='black')

# Final plot settings
plt.axhline(0, color='gray', lw=0.8, ls='--')
plt.xlabel("Party Position $x_P$")
plt.ylabel("Party Utility $U_P$")
plt.title("Strategic Trade-off Between Vote Share and Policy Distance")
plt.legend(loc="upper left", bbox_to_anchor=(0.02, 0.98))
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
